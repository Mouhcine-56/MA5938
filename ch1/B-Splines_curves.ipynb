# B-Splines curves

The B-spline curve in $\mathbb{R}^d$ associated to $T=(t_i)_{0\leqslant i \leqslant n + p + 1}$ and control points $(\mathbf{P}_i)_{ 0 \leqslant i \leqslant n}$ is defined by :
$$
\mathcal{C}(t) = \sum_{i=0}^n N_i^p(t) \textbf{P}_i
$$


## Examples

```python
def example_1():    
    knots  = [0., 0., 0., 1., 1., 1.]
    degree = 2
    n      = len(knots) - degree - 1
    
    P = np.zeros((n, 2))
    P[:, 0] = [1., 1., 0.]
    P[:, 1] = [0., 1., 1.]    
    
    nx = 101
    xs = np.linspace(0., 1., nx)
    
    Q = np.zeros((nx, 2))
    for i,x in enumerate(xs):
        Q[i,:] = point_on_bspline_curve(knots, P, x)

    plt.plot(Q[:,0], Q[:,1], '-b')
    plt.plot(P[:,0], P[:,1], '--ok', linewidth=0.7)
    
    for i in range(0, n):
        x,y = P[i,:]
        plt.text(x+0.05,y+0.05,'$\mathbf{P}_{' + str(i) + '}$')

    plt.axis([-0.2, 1.2, -0.2, 1.2])

    ax = plt.axes()
    ax.set_aspect('equal', 'box')    

example_1()
```
